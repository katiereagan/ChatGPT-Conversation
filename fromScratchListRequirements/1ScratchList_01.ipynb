```python
from PIL import Image, ImageDraw, ImageTk
import tkinter as tk
from tkinter import filedialog

def draw(event):
    x = event.x - canvas_width // 2
    y = canvas_height // 2 - event.y

    if drawing:
        annotation_draw.ellipse((x - marker_size, y - marker_size, x + marker_size, y + marker_size), fill=marker_color, outline=marker_color)

def start_drawing(event):
    global drawing
    drawing = True

def stop_drawing(event):
    global drawing
    drawing = False

def save_annotated_image():
    file_path = filedialog.asksaveasfilename(defaultextension=".jpg")
    if file_path:
        annotated_image.save(file_path)

def save_coordinates():
    file_path = filedialog.asksaveasfilename(defaultextension=".txt")
    if file_path:
        with open(file_path, "w") as f:
            for coordinate in coordinates:
                f.write(f"{coordinate[0]}, {coordinate[1]}\n")

def change_marker_color(color):
    global marker_color
    marker_color = color

def change_marker_size(size):
    global marker_size
    marker_size = size

# Initialize variables
drawing = False
marker_color = "black"
marker_size = 5
coordinates = []

# Create main window
root = tk.Tk()
root.title("Image Annotation")

# Select image from local files
file_path = filedialog.askopenfilename()

# Load and resize the image
image = Image.open(file_path)
resized_image = image.resize((800, 600), Image.LANCZOS)

# Create annotation image and drawing object
annotated_image = image.copy()
annotation_draw = ImageDraw.Draw(annotated_image)

# Create canvas to display the image
canvas_width, canvas_height = resized_image.size
canvas = tk.Canvas(root, width=canvas_width, height=canvas_height)
canvas.pack()

# Convert the resized image to Tkinter-compatible format
photo = ImageTk.PhotoImage(resized_image)

# Display the image on the canvas
annotation_id = canvas.create_image(canvas_width // 2, canvas_height // 2, image=photo)

# Bind mouse events to the canvas
canvas.bind("<B1-Motion>", draw)
canvas.bind("<ButtonPress-1>", start_drawing)
canvas.bind("<ButtonRelease-1>", stop_drawing)

# Create tool buttons
black_marker_button = tk.Button(root, text="Black Marker", command=lambda: change_tool("black_marker"))
black_marker_button.pack(side=tk.LEFT)
white_marker_button = tk.Button(root, text="White Marker", command=lambda: change_tool("white_marker"))
white_marker_button.pack(side=tk.LEFT)
eraser_button = tk.Button(root, text="Eraser", command=lambda: change_tool("eraser"))
eraser_button.pack(side=tk.LEFT)

# Create marker size buttons
size_1_button = tk.Button(root, text="Size 1", command=lambda: change_marker_size(1))
size_1_button.pack(side=tk.LEFT)
size_5_button = tk.Button(root, text="Size 5", command=lambda: change_marker_size(5))
size_5_button.pack(side=tk.LEFT)
size_10_button = tk.Button(root, text="Size 10", command=lambda: change_marker_size(10))
size_10_button.pack(side=tk.LEFT)
size_20_button = tk.Button(root, text="Size 20", command=lambda: change_marker_size(20))
size_20_button.pack(side=tk.LEFT)

# Create eraser size buttons
eraser_5_button = tk.Button(root, text="Eraser 5", command=lambda: change_eraser_size(5))
eraser_5_button.pack(side=tk.LEFT)
eraser_10_button = tk.Button(root, text="Eraser 10", command=lambda: change_eraser_size(10))
eraser_10_button.pack(side=tk.LEFT)
eraser_20_button = tk.Button(root, text="Eraser 20", command=lambda: change_eraser_size(20))
eraser_20_button.pack(side=tk.LEFT)

# Create save buttons
save_annotated_image_button = tk.Button(root, text="Save Annotated Image", command=save_annotated_image)
save_annotated_image_button.pack(side=tk.BOTTOM)
save_coordinate_points_button = tk.Button(root, text="Save Coordinate Points", command=save_coordinate_points)
save_coordinate_points_button.pack(side=tk.BOTTOM)

# Create exit button
exit_button = tk.Button(root, text="Exit", command=exit_program)
exit_button.pack(side=tk.BOTTOM)

# Start the Tkinter event loop
root.mainloop()
```

Here's the modified code that includes variable marker and eraser sizes, a text entry for specific size input, 
highlighting the selected tool in the toolbar, zoom in/out functionality, and an annotation lasso tool:

```python
from PIL import Image, ImageDraw, ImageTk
import tkinter as tk
from tkinter import filedialog, messagebox

# Initialize variables
drawing = False
selected_tool = "black_marker"
marker_size = 5
eraser_size = 5
lasso_active = False
annotation_points = []
black_marker_coordinates = []
white_marker_coordinates = []
erase_coordinates = []

# Create main window
root = tk.Tk()
root.title("Image Annotation")

def draw(event):
    x = event.x - canvas_width // 2
    y = canvas_height // 2 - event.y

    if lasso_active:
        annotation_points.append((x, y))
    elif drawing:
        if selected_tool == "black_marker":
            annotation_draw.ellipse((x - marker_size, y - marker_size, x + marker_size, y + marker_size), fill="black")
            black_marker_coordinates.append((x, y))
        elif selected_tool == "white_marker":
            annotation_draw.ellipse((x - marker_size, y - marker_size, x + marker_size, y + marker_size), fill="white")
            white_marker_coordinates.append((x, y))
        elif selected_tool == "eraser":
            annotation_draw.rectangle((x - eraser_size, y - eraser_size, x + eraser_size, y + eraser_size), fill="white")
            erase_coordinates.append((x, y))

        update_annotation_canvas()

def start_drawing(event):
    global drawing
    drawing = True

def stop_drawing(event):
    global drawing
    drawing = False

def change_tool(tool):
    global selected_tool, lasso_active
    selected_tool = tool
    lasso_active = False
    black_marker_button.config(bg="SystemButtonFace")
    white_marker_button.config(bg="SystemButtonFace")
    eraser_button.config(bg="SystemButtonFace")
    lasso_button.config(bg="SystemButtonFace")

    if tool == "black_marker":
        black_marker_button.config(bg="green")
    elif tool == "white_marker":
        white_marker_button.config(bg="green")
    elif tool == "eraser":
        eraser_button.config(bg="green")
    elif tool == "lasso":
        lasso_button.config(bg="green")
        lasso_active = True

def change_marker_size(size):
    global marker_size
    marker_size = size

def change_eraser_size(size):
    global eraser_size
    eraser_size = size

def update_annotation_canvas():
    photo_annotation = ImageTk.PhotoImage(annotation_image)
    canvas.itemconfigure(annotation_id, image=photo_annotation)

def save_annotated_image():
    file_path = filedialog.asksaveasfilename(defaultextension=".jpg")
    if file_path:
        annotated_image.save(file_path)

def save_coordinate_points():
    file_path = filedialog.asksaveasfilename(defaultextension=".txt")
    if file_path:
        with open(file_path, "w") as f:
            f.write("Black Marker Coordinates:\n")
            for coordinate in black_marker_coordinates:
                f.write(f"{coordinate[0]}, {coordinate[1]}\n")

            f.write("\nWhite Marker Coordinates:\n")
            for coordinate in white_marker_coordinates:
                f.write(f"{coordinate[0]}, {coordinate[1]}\n")

def exit_program():
    if messagebox.askokcancel("Exit", "Are you sure you want to exit?"):
        root.destroy()

def zoom_in():
    canvas.scale("all", canvas_width // 2, canvas_height // 2, 1.2, 1.2)

def zoom_out():
    canvas.scale("all", canvas_width // 2, canvas_height // 2, 0.8, 0.8)

# Create annotation canvas
canvas_width = 800
canvas_height = 600
annotation_image = Image.new("RGB", (canvas_width, canvas_height), "white")
annotation_draw = ImageDraw.Draw(annotation_image)

canvas = tk.Canvas(root, width=canvas_width, height=canvas_height)
canvas.pack()

annotation_id = canvas.create_image(canvas_width // 2, canvas_height // 2, image=None)

canvas.bind("<B1-Motion>", draw)
canvas.bind("<Button-1>", start_drawing)
canvas.bind("<ButtonRelease-1>", stop_drawing)

# Create toolbar
toolbar = tk.Frame(root)
toolbar.pack(pady=10)

black_marker_button = tk.Button(toolbar, text="Black Marker", command=lambda: change_tool("black_marker"))
black_marker_button.pack(side=tk.LEFT, padx=5)
white_marker_button = tk.Button(toolbar, text="White Marker", command=lambda: change_tool("white_marker"))
white_marker_button.pack(side=tk.LEFT, padx=5)
eraser_button = tk.Button(toolbar, text="Eraser", command=lambda: change_tool("eraser"))
eraser_button.pack(side=tk.LEFT, padx=5)
lasso_button = tk.Button(toolbar, text="Lasso", command=lambda: change_tool("lasso"))
lasso_button.pack(side=tk.LEFT, padx=5)

marker_size_scale = tk.Scale(toolbar, label="Marker Size", from_=1, to=20, orient=tk.HORIZONTAL, command=change_marker_size)
marker_size_scale.set(marker_size)
marker_size_scale.pack(side=tk.LEFT, padx=5)

eraser_size_scale = tk.Scale(toolbar, label="Eraser Size", from_=1, to=20, orient=tk.HORIZONTAL, command=change_eraser_size)
eraser_size_scale.set(eraser_size)
eraser_size_scale.pack(side=tk.LEFT, padx=5)

zoom_in_button = tk.Button(toolbar, text="Zoom In", command=zoom_in)
zoom_in_button.pack(side=tk.LEFT, padx=5)
zoom_out_button = tk.Button(toolbar, text="Zoom Out", command=zoom_out)
zoom_out_button.pack(side=tk.LEFT, padx=5)

save_image_button = tk.Button(toolbar, text="Save Image", command=save_annotated_image)
save_image_button.pack(side=tk.LEFT, padx=5)
save_coordinates_button = tk.Button(toolbar, text="Save Coordinates", command=save_coordinate_points)
save_coordinates_button.pack(side=tk.LEFT, padx=5)

exit_button = tk.Button(toolbar, text="Exit", command=exit_program)
exit_button.pack(side=tk.LEFT, padx=5)

root.mainloop()
```
